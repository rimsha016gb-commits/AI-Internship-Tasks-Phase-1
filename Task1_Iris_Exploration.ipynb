## Importing Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset

The Iris dataset is loaded using Seaborn's built-in dataset loader.

```python
# Load Iris dataset via seaborn into a pandas DataFrame
df = sns.load_dataset('iris')

df.head()
```

## Dataset Inspection

Checking dataset shape, columns and basic information.

```python
print("Shape:", df.shape)
print(df.columns.tolist())
df.info()
```

## Data Preprocessing

Checking for missing values and duplicate records.

```python
# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Check duplicates
print("\nDuplicates:", df.duplicated().sum())

# Drop duplicates if any
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

## Statistical Summary

```python
df.describe()
```

## Scatter Plots

Scatter plots help identify relationships between features colored by species.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Scatter Plots - Feature Relationships', fontsize=14)

# Sepal scatter plot
sns.scatterplot(data=df, x='sepal_length', y='sepal_width',
                hue='species', ax=axes[0])
axes[0].set_title('Sepal Length vs Sepal Width')

# Petal scatter plot
sns.scatterplot(data=df, x='petal_length', y='petal_width',
                hue='species', ax=axes[1])
axes[1].set_title('Petal Length vs Petal Width')

plt.tight_layout()
plt.show()
```

## Histograms

Histograms show feature value distributions for each species.

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
fig.suptitle('Histograms - Value Distributions', fontsize=14)

for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    sns.histplot(data=df, x=feature, hue='species', kde=True, ax=axes[row][col])
    axes[row][col].set_title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

## Box Plots

Box plots help detect outliers and compare distributions across species.

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 7))
fig.suptitle('Box Plots - Outlier Detection', fontsize=14)

for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    sns.boxplot(data=df, x='species', y=feature, ax=axes[row][col])
    axes[row][col].set_title(f'{feature} by Species')

plt.tight_layout()
plt.show()
```

# Final Insights

### Observations

1. Setosa species is clearly separated from the other two species based on petal measurements.
2. Petal length and petal width are the strongest distinguishing features among species.
3. Sepal measurements show some overlap among Versicolor and Virginica species.
